```python
# 3. è¨­å®šè¦ä½¿ç”¨çš„æ¨¡å‹
# æ›´å¤šæ¨¡å‹è«‹è¦‹ https://console.groq.com/docs/models

model = "groq:openai/gpt-oss-120b"
```

```python
# 4. LLM å‡ºä»»å‹™
# ä»¥è¬›ç¬‘è©±ç‚ºç¯„ä¾‹

messages = [
    {"role": "system", "content": "è«‹ç”¨å°ç£ç¿’æ…£çš„ä¸­æ–‡å›è¦†ã€‚"}, # system prompt
    {"role": "user", "content": "è·Ÿæˆ‘èªªä¸€å€‹ç¬‘è©±å§ï¼"}, # user prompt
]

response = client.chat.completions.create(
    model=model,
    messages=messages,
)
print(response.choices[0].message.content)
```

```python
# 5. è¨­å®šæç¤ºè© (Prompts)

def generate_story(condition: str) -> str:
    prompt = f"""
è«‹è©¦è‘—æƒ³åƒå‡ºå¦ä¸€ç¨®ã€Šå†°èˆ‡ç«ä¹‹æ­Œã€‹çš„çµå±€ï¼Œæ¢ä»¶å¦‚ä¸‹ï¼š
{condition}

è«‹ä»¥å°èªªæ•˜äº‹æ–¹å¼ï¼Œæ·±åˆ»æç¹ªå„è§’è‰²çš„å‘½é‹èˆ‡æƒ…æ„Ÿï¼Œä½†åƒ…éœ€ä»¥ç°¡çŸ­æ®µè½å‘ˆç¾ã€‚"""

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "ä½ æ˜¯è³‡æ·±å°èªªå®¶ï¼Œæ“…é•·å²è©©å¥‡å¹»é¢¨æ ¼"},
            {"role": "user", "content": prompt},
        ],
        max_tokens=800,
    )
    return response.choices[0].message.content
```

```python
# Gradio ä»‹é¢
with gr.Blocks() as demo:
    gr.Markdown("## â„ï¸ğŸ”¥ã€Šå†°èˆ‡ç«ä¹‹æ­Œã€‹çµå±€é‡å¯«å™¨")
    gr.Markdown("è¼¸å…¥æ¢ä»¶ï¼Œä¾‹å¦‚ï¼šã€Œé¾å¥³æ´»è‘—ä½†ç“Šæ©æ­»äº†ã€ã€ã€Œå¤œç‹å‹å‡ºï¼Œé¾å¥³å’Œç“Šæ©éƒ½æ­»äº†ã€ï¼Œæˆ–å…¶ä»–å‰µæ„è¨­å®šã€‚")

    condition_input = gr.Textbox(
        label="çµå±€æ¢ä»¶",
        placeholder="ä¾‹å¦‚ï¼šé¾å¥³æ´»è‘—ä½†ç“Šæ©æ­»äº†",
    )
    generate_btn = gr.Button("é»æˆ‘ç”Ÿæˆæ–°çµå±€")
    story_output = gr.Markdown(label="é‡å¯«çš„çµå±€")

    generate_btn.click(fn=generate_story, inputs=condition_input, outputs=story_output)
```
